# **Costomizacion de graficas**

```python
import matplotlib.pyplot as plt
```

```python
year=[1940,1950,1960,1970,1980,1990,2000,2010,2020]
pop=[2.538,2.7,2.9,3.692,4.6,5.263,5.91,6.972,7.120]

year=[1800,1850,1900]+year
pop=[1.0,1.262,1.650]+pop
```

```python
plt.plot(year,pop)
plt.ylabel("Poblacion")
plt.xlabel("anio")
plt.title("Proyeccion de Ploblacion Mundial")
tick_val=[0,2,4,6,8]
plt.yticks(tick_val,   #Aqui sirve para determinar los valores a ver en la linea  y asignarles un Alias
           ['0','2B','4B','6B','8B'],)
```
